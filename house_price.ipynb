# **Data Visualization**

```python
df.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
columns_to_plot = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

fig,axes = plt.subplots(1,len(columns_to_plot),figsize=(32,7))

for ax,col in zip(axes,columns_to_plot):
  sns.countplot(data=df,x=col,ax=ax)
  ax.set_title(f"count plot of {col}")
  ax.set_xlabel(col)
  ax.set_ylabel("count")
plt.tight_layout()
plt.show()
```
